# IDENTIFY DATA TYPES PROPERLY

- Numerical (continuous): Can take any value, mathematical operations meaningful
- Numerical (discrete): Counts, integers
- Ordinal: Ordered categories, but intervals not equal
- Nominal: Unordered categories
- Binary: Special case of nominal with 2 categories

### Function to classify columns into proper data types

```python
def classify_columns(df):
    classification = {
        'numerical_continuous': [],
        'numerical_discrete': [],
        'ordinal': [],
        'nominal': [],
        'binary': [],
        'id': [],
        'target': []
    }
    
    for col in df.columns:
        # Skip target for now
        if col == 'Asthma':
            classification['target'].append(col)
            continue
        
        # Check if it's an ID column
        if col == 'Patient_ID' or 'id' in col.lower():
            classification['id'].append(col)
            continue
        
        # Check data type
        if df[col].dtype in ['int64', 'float64']:
            unique_count = df[col].nunique()
            
            # If only 2 unique values, it's binary
            if unique_count == 2:
                classification['binary'].append(col)
            # If less than 20 unique values and integers, could be discrete
            elif unique_count < 20 and df[col].dtype == 'int64':
                classification['numerical_discrete'].append(col)
            else:
                classification['numerical_continuous'].append(col)
        else:
            # Categorical columns
            unique_count = df[col].nunique()
            
            # Check if it's ordinal based on known patterns
            ordinal_patterns = ['level', 'status', 'rating', 'pollution', 'activity']
            if any(pattern in col.lower() for pattern in ordinal_patterns):
                classification['ordinal'].append(col)
            else:
                classification['nominal'].append(col)
    
    return classification
```

```python
col_classification = classify_columns(df)
```

### Re-classify columns with new features

```python
def update_classification(df):
    """
    Update column classification including new features
    """
    classification = {
        'numerical_continuous': [],
        'numerical_discrete': [],
        'ordinal': [],
        'nominal': [],
        'binary': []
    }
    
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            unique_count = df[col].nunique()
            
            if unique_count == 2:
                classification['binary'].append(col)
            elif unique_count < 20 and df[col].dtype == 'int64':
                classification['numerical_discrete'].append(col)
            else:
                classification['numerical_continuous'].append(col)
        else:
            # Check if ordinal based on name
            ordinal_patterns = ['level', 'status', 'category', 'group']
            if any(pattern in col.lower() for pattern in ordinal_patterns):
                classification['ordinal'].append(col)
            else:
                classification['nominal'].append(col)
    
    return classification
```

# CREATE PREPROCESSING PIPELINE (CRITICAL STEP)

```python
print("""
This is where we apply different treatments to different data types:
- Numerical: Scale to mean=0, std=1
- Ordinal: Encode with order, then scale
- Nominal: One-hot encode
- Binary: Keep as is (already 0/1)
""")
```

### Pipeline Components

```python
print(f"  • Numerical ({len(numerical_cols)}): {numerical_cols}")
print(f"  • Ordinal ({len(ordinal_cols)}): {ordinal_cols}")
print(f"  • Nominal ({len(nominal_cols)}): {nominal_cols}")
print(f"  • Binary ({len(binary_cols)}): {binary_cols[:5]}...")
```

### IMPORTANT: First, check the actual values in data

```python
for col in ordinal_cols:
    print(f"\n{col} unique values in data: {sorted(df_fe[col].unique())}")
```

###  Define ordinal categories order

```python
ordinal_categories = []

for col in ordinal_cols:
    if col == 'Smoking_Status':
        # Use the actual values from your data
        actual_values = sorted(df_fe[col].unique())
        # Ensure correct order (Non Smoker < Ex Smoker < Current Smoker)
        if 'Current Smoker' in actual_values and 'Ex Smoker' in actual_values and 'Non Smoker' in actual_values:
            ordinal_categories.append(['Non Smoker', 'Ex Smoker', 'Current Smoker'])
        else:
            # Fallback to whatever order exists
            ordinal_categories.append(actual_values)
        print(f"  • Smoking_Status order: {ordinal_categories[-1]}")
    
    elif col == 'Air_Pollution_Level':
        ordinal_categories.append(['Low', 'Moderate', 'High'])
        print(f"  • Air_Pollution_Level order: {ordinal_categories[-1]}")
    
    elif col == 'Physical_Activity_Level':
        ordinal_categories.append(['Low', 'Moderate', 'High'])
        print(f"  • Physical_Activity_Level order: {ordinal_categories[-1]}")
    
    elif col == 'BMI_Category':
        ordinal_categories.append(['Underweight', 'Normal', 'Overweight', 'Obese'])
        print(f"  • BMI_Category order: {ordinal_categories[-1]}")
    
    elif col == 'Age_Group':
        ordinal_categories.append(['Child', 'Teenager', 'Young_Adult', 'Adult', 'Senior'])
        print(f"  • Age_Group order: {ordinal_categories[-1]}")
    
    else:
        # For any other ordinal columns, use the actual sorted values
        ordinal_categories.append('auto')
        print(f"  • {col}: using auto order")
```

## Create transformers for each data type

```python
transformers = []
```

### 1. Numerical transformer

```python
if numerical_cols:
    numerical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    transformers.append(('numerical', numerical_transformer, numerical_cols))
    print(f"\n✅ Added 'numerical' transformer for {len(numerical_cols)} columns")
```

### 2. Ordinal transformer

```python
if ordinal_cols:
    ordinal_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ordinal', OrdinalEncoder(categories=ordinal_categories)),
        ('scaler', StandardScaler())
    ])
    transformers.append(('ordinal', ordinal_transformer, ordinal_cols))
    print(f"✅ Added 'ordinal' transformer for {len(ordinal_cols)} columns")
```

### 3. Nominal transformer

```python
if nominal_cols:
    nominal_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
        ('onehot', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
    ])
    transformers.append(('nominal', nominal_transformer, nominal_cols))
    print(f"✅ Added 'nominal' transformer for {len(nominal_cols)} columns")
```

### 4. Binary transformer (passthrough)

```python
if binary_cols:
    transformers.append(('binary', 'passthrough', binary_cols))
    print(f"✅ Added 'binary' transformer for {len(binary_cols)} columns")
```

## Verify transformer names are unique

### Define models to train

```python
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced'),
    'Decision Tree': DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'XGBoost': xgb.XGBClassifier(n_estimators=100, random_state=42, eval_metric='logloss'),
    'SVM': SVC(random_state=42, class_weight='balanced', probability=True),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB()
}
```
