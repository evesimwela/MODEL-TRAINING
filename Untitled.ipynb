## Set up the plotting style

## Plot distributions for numeric columns

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

numeric_cols = ['Age', 'BMI']
for i, col in enumerate(numeric_cols):
    axes[i].hist(df[col], bins=30, edgecolor='black', alpha=0.7, color='steelblue')
    axes[i].set_title(f'Distribution of {col}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)
    mean_val = df[col].mean()
    median_val = df[col].median()
    std_val = df[col].std()
    
    axes[i].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.1f}')
    axes[i].axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.1f}')
    axes[i].axvline(mean_val + std_val, color='orange', linestyle=':', linewidth=1.5, alpha=0.7, label=f'+1 Std: {mean_val+std_val:.1f}')
    axes[i].axvline(mean_val - std_val, color='orange', linestyle=':', linewidth=1.5, alpha=0.7, label=f'-1 Std: {mean_val-std_val:.1f}')
    
    axes[i].legend(fontsize=9)
    axes[i].grid(True, alpha=0.3)
    
    stats_text = f'Min: {df[col].min():.1f}\nMax: {df[col].max():.1f}\nStd: {std_val:.1f}'
    axes[i].text(0.02, 0.98, stats_text, transform=axes[i].transAxes, 
                fontsize=9, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.suptitle('Numeric Features Distribution', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Plot count plots for categorical columns

```python
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
axes = axes.ravel()  # Flatten the 2D array of axes

categorical_cols_for_plot = ['Gender', 'Smoking_Status', 'Asthma', 'HistoryOfAllergies', 
                            'Wheezing', 'ShortnessOfBreath', 'ChestTightness', 'Coughing', 'Diagnosis']

for i, col in enumerate(categorical_cols_for_plot, start=0):  # Start from 0 to use first 9 axes
    if i < len(axes):
        # Get value counts
        value_counts = df[col].value_counts().sort_index()
        
        # Create bar plot
        value_counts.plot(kind='bar', ax=axes[i], color=['skyblue', 'salmon'])
        
        # Customize the plot
        axes[i].set_title(f'{col} Distribution', fontsize=12, fontweight='bold')
        axes[i].set_xlabel(col, fontsize=10)
        axes[i].set_ylabel('Count', fontsize=10)
        
        # Rotate x-axis labels if needed
        axes[i].tick_params(axis='x', rotation=45)
        
        # Add value labels on top of bars
        for j, v in enumerate(value_counts.values):
            axes[i].text(j, v + 0.5, str(v), ha='center', va='bottom', fontsize=9)

# Remove any empty subplots
for i in range(len(categorical_cols_for_plot), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

## Correlation Analysis

```python
correlation_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Features')
plt.tight_layout()
plt.show()
```

```python
target_correlations = correlation_matrix['Diagnosis'].sort_values(ascending=False)
print("\nCorrelation with Diagnosis (Target):")
print(target_correlations)

high_corr = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            high_corr.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))

print("\nHighly Correlated Feature Pairs (|r| > 0.7):")
if high_corr:
    for pair in high_corr:
        print(f"{pair[0]} - {pair[1]}: {pair[2]:.3f}")
else:
    print("None found")
```

## Analyze Feature Relationships with Target

. Independent t-test → for continuous variables (Age, BMI)

. Chi-square test → for categorical variables

Graphical bars - categorical

Graphical Box plots for numeric features

```python
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
axes = axes.ravel()

features = ['Age', 'BMI', 'Gender', 'Smoking_Status', 'Asthma', 'HistoryOfAllergies',
            'Wheezing', 'ShortnessOfBreath', 'ChestTightness', 'Coughing']

for i, feature in enumerate(features):
    if feature in ['Age', 'BMI']:
        # Box plots for numeric features
        df.boxplot(column=feature, by='Diagnosis', ax=axes[i])
        axes[i].set_title(f'{feature} by Diagnosis')
        axes[i].set_xlabel('Diagnosis (0=No Asthma, 1=Asthma)')
    else:
        # Grouped bar charts for categorical features
        pd.crosstab(df[feature], df['Diagnosis']).plot(kind='bar', ax=axes[i], color=['skyblue', 'salmon'])
        axes[i].set_title(f'{feature} by Diagnosis')
        axes[i].set_xlabel(feature)
        axes[i].set_ylabel('Count')
        axes[i].legend(['No Asthma', 'Asthma'])

plt.tight_layout()
plt.show()
```

### T-test for numeric features

```python
from scipy.stats import chi2_contingency, ttest_ind
for feature in ['Age', 'BMI']:
    asthma_group = df[df['Diagnosis'] == 1][feature]
    no_asthma_group = df[df['Diagnosis'] == 0][feature]
    t_stat, p_value = ttest_ind(asthma_group, no_asthma_group)
    print(f"{feature}: t-statistic={t_stat:.3f}, p-value={p_value:.4f}")
```

```python

# Chi-square test for categorical features
for feature in ['Gender', 'Smoking_Status', 'Asthma', 'HistoryOfAllergies',
                'Wheezing', 'ShortnessOfBreath', 'ChestTightness', 'Coughing']:
    contingency_table = pd.crosstab(df[feature], df['Diagnosis'])
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    print(f"{feature}: chi2={chi2:.3f}, p-value={p_value:.4f}")
```
